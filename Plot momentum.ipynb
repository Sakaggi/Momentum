```
import matplotlib.pyplot as plt

#ax = plt.scatter(df['D'], df['B'], s=df['C']*12000, c="red", alpha=0.7)
MEDIUM_SIZE = 25
SMALL_SIZE = 15
#df.Game=df.Game-df.Gamecou
ax = df.plot.scatter(x='D', y='B', s=df['Momentum']*12000, c="red", alpha=0.5)
plt.xlabel("Game Length")
plt.ylabel("Game Information")
#plt.title("Momentum")
plt.rc('axes', labelsize=SMALL_SIZE)
plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)
for i, txt in enumerate(df['Game']):
    ax.annotate(txt, (df.D.iat[i],df.B.iat[i]), fontsize=14)
#plt.show()
#set_fontsize(10)
fig = plt.gcf()
fig.set_size_inches(18.5, 14.5)
fig.savefig('test2png.png', dpi=75)

from matplotlib2tikz import save as tikz_save
tikz_save("/content/drive/My Drive/test.tex")
```
